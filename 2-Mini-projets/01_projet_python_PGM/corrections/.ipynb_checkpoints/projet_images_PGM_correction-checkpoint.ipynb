# Traitement d'images au format PGM

## Introduction

Une image peut tre stocke sous la forme d'un tableau de points lmentaires appels pixels (abrviation de picture elements). On s'intressera dans un premier temps  la manipulation d'images
 en niveaux de gris (ou grayscale en anglais), cest--dire, des images dont les ''couleurs'' sont uniquement des nuances de gris. On peut reprsenter une telle image par une matrice dentiers, dont la valeur des lments reprsente lintensit lumineuse des pixels de limage. Par consquent, un traitement dimage peut tre ralis en manipulant 
 la matrice qui la reprsente.

## Fichier PGM

Un fichier PGM (Portable Gray Map) est compos de deux parties : 

* Un entte indiquant les paramtres de l'image
* Une matrice dont chaque lment reprsente le niveau de gris dun pixel de limage.

 Lentte contient 3 lignes. La premire ligne est toujours la mme, il sagit de lidentifiant du fichier PGM : P2 pour une image code en ASCII et P5 pour une image code en binaire. La seconde ligne contient 2 entiers spars par un espace: le nombre de colonnes et le nombre de lignes de pixels de limage.

  



 Enfin, la troisime ligne contient la valeur maximale que peut prendre un pixel (par exemple : 255 pour un codage des niveaux de gris entre 0 et 255). 
 Les lignes qui suivent donnent lintensit de chaque pixel. Ceux-ci sont numrs ligne par ligne et colonne par colonne. Ainsi la premire valeur est lintensit du pixel du coin suprieur gauche, la seconde est lintensit du pixel de la premire ligne, deuxime colonne, etc.

Par exemple, un fichier PGM contenant les informations ci-dessous est lencodage dune image reprsentant une barre oblique blanche sur fond noir (voir Fig. 1 du sujet en pdf).

* crer une fonction fun(x, y) qui  partir d'un couple de rels (x, y) 
retourne l'valuation de $\frac{1}{2} (\sin(x)\; \sin(y)+1)$.
La valeur retourne est dans $[0, 1]$.

```python
#CORRECTION
def fun(x, y):
    """ f(x,y)=0.5*(sin(x)*sin(y)+1) in [0, 1]  """
    return 0.5*(np.sin(x)*np.sin(y)+1.0)
```

* crer une fonction centering(i, L) qui  partir de deux entiers $i$ et $L$
retourne $2 i/L-1$. La fonction vrifiera que $i \in [0, L[$. 

On pourra mettre en oeuvre l'instruction suivante : 
assert condition, "out of range" en adaptant la condition  votre usage.

```python
#CORRECTION
def centering(i, L):
    return (2.0*i-L)/L
```

* crer une fonction createImg(fun, NROWS=MAXTAIL, NCOLS=MAXTAIL, NLEVELS = MAXDEG)
qui retourne un tableau img de type numpy de taille NROWS $\times$ NCOLS contenant les valeurs
des pixels comprises entre 0 et 255. 

Pour tout couple d'indice $(i, j) \in $ [0, NROWS[ $\times$ [0, NCOLS[, dfinir
x=$\pi$ centering(i, NROWS) et y=$\pi$  centering(j, NCOLS)
et stocker dans l'lment $(i, j)$ la valeur de NLEVELS $\times$ fun(x, y)
dont on ne gardera que la partie entire. 

Pour ce faire utiliser la fonction $floor$
du module math.



```python
#CORRECTION
def createImg(fun, NROWS = MAXTAIL, NCOLS = MAXTAIL, NLEVELS = MAXDEG):   
    img = np.zeros(shape=(NROWS, NCOLS), dtype='uint8')
    for i in range(NROWS):
        x = centering(i, NROWS)*np.pi
        for j in range(NCOLS):
            y = centering(j, NCOLS)*np.pi
            img[i, j]=math.floor(fun(x,y)*NLEVELS)
    return img
```

* crer une fonction writeImg(img, filename) qui sauvegarde
dans le fichier filename l'image img au format PGM ASCII.



```python
#CORRECTION
def writeImg(img, filename):
    nrows, ncols = img.shape
    nlevs = img.max()
    
    with open(filename, 'w') as f:
        f.write('P2\n')
        f.write(str(ncols)+' '+str(nrows)+'\n')
        f.write(str(nlevs)+'\n')
        for row in img:
            for e in row:
                #chaine = "{var:4d}".format(var=e)
                chaine = f"{e:4d}"            
                f.write(chaine+' ')
            f.write('\n')
```

```python
# CODE FOURNI AUX APPRENANTS
def readImg(filename):
    with open(filename) as f:
        lines = f.readlines()

    # Ignores commented lines
    for l in list(lines):
        if l[0] == '#':
            lines.remove(l)

    # Makes sure it is ASCII format (P2)
    # leading and trailing whitespace removed
    assert lines[0].strip() == 'P2' 

    # Converts data to a list of integers
    data = []
    for line in lines[1:]:
        data.extend([int(c) for c in line.split()])

    return (np.array(data[3:]), (data[1],data[0]), data[2])
```

* Crer une fonction createImgVec(funvec, NROWS = MAXTAIL, NCOLS = MAXTAIL, NLEVELS = MAXDEG)
qui utilisera la fonction numpy np.meshgrid(x, y) o
x est un tableau numpy de NROWS valeurs qui-espaces entre $[-\pi,\pi]$
et y de NCOLS valeurs qui-espaces entre $[-\pi,\pi]$.
La fonction funvec mettra en oeuvre les versions vectorielles numpy de
la fonction sinus. 

La documentation de np.floor nous informe que la valeur retourne est un rel.
L'image tant code avec des entiers non signs 8 bits,
il faut donc convertir le type en utilisant la mthode astype, vue dans la
section 3.

```python
def createImgVec(fun, NROWS, NCOLS, NLEVELS=255):    
    x = np.arange(0, NROWS)
    x = centering(x, NROWS)*np.pi
    y = np.arange(0, NCOLS)
    y = centering(y, NCOLS)*np.pi
    X, Y = np.meshgrid(x, y)
    A=np.floor(fun(X, Y)*NLEVELS).astype(np.uint8)
    return A
```

# BONUS : stockage des images en format binaire P5

On dsire modifier la fonction writeImg(img, filename) pour stocker limage en niveau
de gris dans un fichier au format binaire P5.

L'ouverture d'un fichier binaire en criture se fait avec la fonction python open(filename, wb)
et en lecture par open(filename, rb).
Noter que la conversion d'un tableau d'entiers z en octets peut tre ralise avec la fonction membre z.tobytes().

On vous donne les routines d'criture et de lecture suivantes.

## Exercice :
Vrifier leur bon fonctionnemment.

```python
# Version efficace avec numpy
import numpy as np

def writeImg(img, filename):    
    """
    Ecrit une image contenue dans un tableau numpy dans un fichier PGM format P5 (binaire)
    """
    nrows, ncols = img.shape
    nlevs = img.max()
    assert nlevs < 256, f"Le nombre de niveaux de gris ({nlevs}) dpasse 255"
    
    with open(filename, 'wb') as f:
        # criture de l'en-tte en ASCII
        # f.write(...) en mode binaire ('wb') attend des octets (bytes), pas une chane
        f.write(b'P5\n')
        f.write(f'{ncols} {nrows}\n'.encode('ascii'))
        f.write(f'{nlevs}\n'.encode('ascii'))
        
        # Conversion en uint8 et criture directe
        img_bytes = img.astype(np.uint8).tobytes()
        f.write(img_bytes)
```

## Lecture d'un fichier au format binaire 'P5'

```python
import struct
import numpy as np

import math
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from PIL import Image
from time import process_time

def readImg(filename):
    """
    Lit un fichier PGM format P5 (binaire)
    Retourne l'image sous forme de tableau numpy
    """
    with open(filename, 'rb') as f:
        # Lecture de l'en-tte
        # La mthode .strip() en Python supprime les espaces et caractres invisibles au dbut et  la fin d'une chane de caractres.
        magic = f.readline().decode('ascii').strip()
        if magic != 'P5':
            raise ValueError(f"Format non support: {magic}. Attendu: P5")
        
        # Lecture des dimensions (en grant les commentaires)
        line = f.readline().decode('ascii').strip()
        while line.startswith('#'):
            line = f.readline().decode('ascii').strip()

        # Ecriture compacte de
        # tokens = line.split(); ncols, nrows = int(tokens[0]), int(tokens[1])
        ncols, nrows = map(int, line.split())
        
        # Lecture de la valeur maximale
        line = f.readline().decode('ascii').strip()
        while line.startswith('#'):
            line = f.readline().decode('ascii').strip()
        
        max_val = int(line)
        
        # Lecture de nrows x ncols octets
        data = f.read(nrows * ncols)

        # np.frombuffer(...) cre un tableau NumPy sans copier les donnes, directement depuis le buffer mmoire (data)
        img = np.frombuffer(data, dtype=np.uint8)
        # Reshape en matrice 2D
        img = img.reshape(nrows, ncols)
        
    return img
    
```
