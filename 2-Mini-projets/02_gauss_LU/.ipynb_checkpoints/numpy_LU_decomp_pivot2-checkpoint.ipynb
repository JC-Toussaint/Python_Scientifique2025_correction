```python
# A : matrice a factoriser de dim nxn
import numpy as np
np.set_printoptions(precision=4, suppress=True)

A=np.array([[-4, 2, 1], [5, 2, 1], [5, -6, 2]])    
n=A.shape[0]
Id=np.eye(n, n); # Matrice Identite
L=Id;

for k in range(n-1):
# Construction de la matrice d’elimination
    inv_Lk=np.eye(n, n)
    for i in range(k+1, n):
        inv_Lk[i,k]=-A[i,k]/A[k,k];

    print(inv_Lk)
# Construction de la matrice A(k+1) a partir de A(k) 
    A=inv_Lk.dot(A)
    Lk=-inv_Lk+2*np.eye(n, n) # Inverse de la matrice d’elimination
    L=L.dot(Lk);

# matrice A(k+1) est la matrice triangulaire superieure
U=A
print('---------')
print(U)
print(L)
print(L.dot(U))
```

```python
# A : matrice a factoriser de dim nxn
import numpy as np
np.set_printoptions(precision=4, suppress=True)

A=np.array([[-4, 2, 1, 1], [5, 2, 1, 2], [5, -6, 2, 2], [7, 8, 2, 3]])  
n=A.shape[0]
L=np.eye(n, n) # Matrice Identite
P   =np.eye(n, n)
inv_P=np.eye(n, n)

for k in range(n-1):
# Construction de la matrice de permutation
    Pk    = np.eye(n, n)
    colk=np.abs(A[:, k])
    kmax=colk.argmax()
    if kmax>k:
        Pk[[k, kmax]]=Pk[[kmax, k]]
    A=Pk.dot(A)
    
# Construction de la matrice d’elimination
    inv_Lk= np.eye(n, n)
    for i in range(k+1, n):
        inv_Lk[i,k]=-A[i,k]/A[k,k];

# Construction de la matrice A(k+1) a partir de A(k) 
    A=inv_Lk.dot(A)
    Lk=-inv_Lk+2*np.eye(n, n) # Inverse de la matrice d’elimination
    Lk=Pk.dot(Lk)
    L=L.dot(Lk)
    P=P.dot(Pk) # Pk @ Pk = Id

# matrice A(k+1) est la matrice triangulaire superieure
U=A
print('inv_P ---------')
print(inv_P)
print('P ---------')
print(P)

inv_P=P.T
print('Id----------')
print(inv_P @ P)
print('-----------')
L=inv_P.dot(L)
print(P)
print('L----------')
print(L)
print(U)
print(P @ (L @ U))
```
