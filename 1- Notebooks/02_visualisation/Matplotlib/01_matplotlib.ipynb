# Matplotlib

Sa vocation est de produire des figures de qualité en utilisant des fonctions
qui s'apparentent à celles de Matlab.
De nombreux formats d'exportation existent et permettent de produire
des documents publiables.
Cette librairie a été portée sous un ensemble d'environnements.

## simple plot(x,y)
On désire tracer les valeurs d'un tableau y en fonction de celles d'un tableau x

```python
%matplotlib inline

# pour afficher dans une fenêtre graphique indépendante
%matplotlib qt  

import matplotlib.pyplot as plt
import numpy as np

plt.close('all') #efface fenêtre graphique

# data
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
y2 = np.cos(x)

plt.plot(x, y, label = 'sin(x²)',color='green', linewidth=1, linestyle='-', marker = '+')
plt.plot(x, y2, label = 'cos(x)', color='red', linewidth=2, linestyle='-', marker = '*')
#plt.grid()
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.title('titre')
plt.legend()
plt.grid()
plt.show()

```

## Assemblage de figures dans une page
On retrouve une fonction similaire à la fonction subplot de Matlab

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, y, color='red', linewidth=2.0, linestyle="-")
plt.title('titre1')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, y, color='blue', linewidth=2.0, linestyle="-", label='sin')
plt.title('titre2')
plt.ylabel('axe y')
plt.legend(loc='best')

plt.subplot(2, 1, 2)
plt.plot(x, y, color='red', linewidth=2.0, linestyle="-")
plt.title('titre3')

plt.tight_layout()  # ← ajustement automatique
plt.show()
```
