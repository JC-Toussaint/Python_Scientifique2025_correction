## Matplotlib
Tracé d'une courbe paramétrique en 3D

```python
%matplotlib inline 

# pour afficher dans une fenêtre graphique indépendante
# %matplotlib qt  

import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 14

fig = plt.figure()
# Méthode moderne pour créer un subplot 3D
ax = fig.add_subplot(111, projection='3d')

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='courbe parametrique')
ax.legend()

plt.show()
```

## Interactivité

Interactivité de Matplotlib dans jupyter difficile. 

```python
# ============================================
# SOLUTION : Plotly 
# ============================================
# Installation : pip install plotly

import plotly.graph_objects as go
import numpy as np

plt.close('all') #efface fenêtre graphique

# Recréer les données
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

# Créer le graphique Plotly
fig_plotly = go.Figure(data=go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='lines',
    name='Courbe paramétrique',
    line=dict(color='blue', width=6)
))

fig_plotly.update_layout(
    title='Courbe 3D Interactive - Faites glisser pour tourner',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=1.5)
        )
    ),
    width=800,
    height=600
)

fig_plotly.show()

```

```python
# ============================================
# SOLUTION : Matplotlib avec vues multiples
# ============================================
%matplotlib inline

plt.close('all') #efface fenêtre graphique
fig = plt.figure(figsize=(15, 5))

# Vue 1
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot(x, y, z, 'b-', linewidth=2)
ax1.set_title('Vue 1')
ax1.view_init(elev=20, azim=45)

# Vue 2
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot(x, y, z, 'r-', linewidth=2)
ax2.set_title('Vue 2')
ax2.view_init(elev=60, azim=120)

# Vue 3
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot(x, y, z, 'g-', linewidth=2)
ax3.set_title('Vue 3')
ax3.view_init(elev=10, azim=180)

plt.tight_layout()
plt.show()

# ============================================
# Fonction pour générer différentes vues
# ============================================

def plot_3d_view(elevation=30, azimuth=45):
    """Fonction pour générer une vue 3D avec angles personnalisés"""
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    ax.plot(x, y, z, label='courbe parametrique', linewidth=2)
    ax.legend()
    ax.set_xlabel('X')
    ax.set_ylabel('Y') 
    ax.set_zlabel('Z')
    ax.set_title(f'Vue 3D (élévation: {elevation}°, azimut: {azimuth}°)')
    
    ax.view_init(elev=elevation, azim=azimuth)
    plt.show()

# Exemples d'utilisation :
# plot_3d_view(20, 45)   # Vue de face
# plot_3d_view(60, 120)  # Vue du dessus
# plot_3d_view(0, 0)     # Vue de côté
```
