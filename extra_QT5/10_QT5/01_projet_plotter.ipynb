```python
# version 2 #
import sys

from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as canevas
from matplotlib.figure import Figure
import numpy as np

from PyQt5.QtCore import Qt, QPoint, QRect
from PyQt5.QtWidgets import QWidget, QApplication, QHBoxLayout, QVBoxLayout, QPushButton, QTableWidget, QTableWidgetItem, QLineEdit, QFileDialog

class fenetre(QWidget):
    """ la fenetre principale """
    def __init__(self):
        super().__init__()
        self.resize(900, 500)
        self.setWindowTitle('plotter v.2')
        
        # definition du canevas
        fig = Figure((4, 4))
        self.can = canevas(fig)
        self.can.setParent(self)
        self.graph = fig.add_subplot(111)
        
        # TableWidget
        self.tab_data=QTableWidget(1000,2)
        self.tab_data.setHorizontalHeaderLabels(['x', 'y'])
        
        # remplissage par une fonction sinus
        x=np.linspace(0,10,1000)
        y=np.sin(x)
        self.fill(x,y)
        self.plot()
        
        # lineEdit xmin, xmax et nb_pts
        self.le_xmin = QLineEdit()
        self.le_xmin.setText('0')
        self.le_xmax = QLineEdit()
        self.le_xmax.setText('10')
        self.le_nbpt = QLineEdit()
        self.le_nbpt.setText('1000')
        lay_x = QHBoxLayout()
        lay_x.addWidget(self.le_xmin)
        lay_x.addWidget(self.le_xmax)
        lay_x.addWidget(self.le_nbpt)
        
        # lineEdit formula et bouton update
        self.le_formula = QLineEdit()
        self.le_formula.setText("np.sin(x)")
        btn_update = QPushButton("update graph")
        btn_update.clicked.connect(self.update)
        btn_update.setShortcut('Return')
        lay_left = QVBoxLayout()
        lay_left.addLayout(lay_x)
        lay_left.addWidget(self.le_formula)
        lay_left.addWidget(btn_update)
        
        # bouton save
        btn_save = QPushButton("save")
        btn_save.clicked.connect(self.save)
        lay_left.addWidget(btn_save)
       
        # insertion du canevas dans un layout et affichage
        lay = QHBoxLayout(self)
        lay.addLayout(lay_left)
        lay.addWidget(self.can)
        lay.addWidget(self.tab_data)
        
        self.show()
    
    def fill(self, x, y):
        self.tab_data.setRowCount(x.shape[0])
        for i in range(x.shape[0]):
            self.tab_data.setItem(i, 0, QTableWidgetItem(str(x[i])))
            self.tab_data.setItem(i, 1, QTableWidgetItem(str(y[i])))
            
    def plot(self):
        x = []
        y = []
        for i in range(self.tab_data.rowCount()):
            x.append(float(self.tab_data.item(i,0).text()))
            y.append(float(self.tab_data.item(i,1).text()))
        self.graph.clear()
        self.graph.plot(x,y)
        self.can.draw()
        
    def update(self):
        # mise à jour de x
        xmin = float(self.le_xmin.text())
        xmax = float(self.le_xmax.text())
        nbpt = int(self.le_nbpt.text())
        x = np.linspace(xmin, xmax, nbpt)
        # mise à jour du QTableWidget
        y = eval(self.le_formula.text())
        self.fill(x,y)
        self.plot()
    
    def save(self):
        options = QFileDialog.Options()
        options |= QFileDialog.DontUseNativeDialog
        nom_fichier = QFileDialog.getSaveFileName(self, "enregister sous :", options=options)
        with open(nom_fichier[0], 'w') as file:
            for i in range(self.tab_data.rowCount()-1):
                x = self.tab_data.item(i,0).text()
                y = self.tab_data.item(i,1).text()
                file.write(x + " " + y + "\n")
    
if __name__=='__main__':
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    f1 = fenetre()
    app.exec_()
```
